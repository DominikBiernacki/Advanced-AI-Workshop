### Więcej o listach

```python
a = [1, 2, 3]

print("Pierwszy element:", a[0])
print("Ostatni element:", a[-1])
print("Wszystkie elementy poza pierwszym:", a[1:])

a.append(4)

print(a)

a.remove(2)

print(a)

print(a.index(3))
```
